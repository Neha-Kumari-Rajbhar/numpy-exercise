```python
import pandas as pd
```

```python
df=pd.DataFrame({
"City":["Delhi","Delhi","Delhi","Mumbai","Mumbai","Mumbai"],
"Product":["Shoes","Shoes","Watch","Shoes","Watch","Watch"],
"Sales":[200,150,300,400,250,300]
})
```

```python
df.pivot_table(values="Sales", index="City", columns="Product", aggfunc="sum",fill_value=0)
```
