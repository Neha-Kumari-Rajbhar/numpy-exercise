**Pandas Series**

A Series is a one-dimensional labeled array capable of holding any data type.  
The axis labels are collectively called the **index**.

**Creating Series**

There are multiple ways to create a pandas Series:


```python
import numpy as np
import pandas as pd
```

```python
lables = ['a','b','c'] #lable
my_list=[1,2,3] #list
arr=np.array([10,20,30]) #array
dic={'a':1 , 'b':2 , 'c':3}
```

```python
pd.Series(my_list)
```

```python
pd.Series(arr)
```

```python
pd.Series(my_list,index=lables)
```

```python
pd.Series(arr,index=my_list)
```

```python
pd.Series(arr,index=dic)
```

```python
pd.Series(dic)
```
