```python
import numpy as np
import pandas as pd
```

**Concatination of 2 dataframes**

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df2,df1])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
df3 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2']
})

df4 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
pd.concat([df4,df3])
```

```python
pd.concat([df3,df4],axis=1)
```

```python
df5 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index=[1, 2, 3])

# Second DataFrame
df6 = pd.DataFrame({
    'score': [85, 90, 75]
}, index=[5, 6, 7])
```

```python
pd.concat([df5,df6])
```

**Merging 2 dataframes**

```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees.merge(salaries,on='employee_id')
```

```python
pd.merge(employees,salaries,on='employee_id')
```

```python
pd.merge(salaries,employees,on='employee_id',how='outer')
```

```python
pd.merge(salaries,employees,on='employee_id',how='left')
```

```python
pd.merge(salaries,employees,on='employee_id',how='right')
```

**Joining 2 data frames**

```python
student = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index=[1, 2, 3])

# Second DataFrame
marks = pd.DataFrame({
    'score': [85, 90, 75]
}, index=[2, 3, 4])
```

```python
student.join(marks)
```

```python
marks.join(student)
```

```python
student.join(marks,how='outer')
```

```python
student.merge(marks, left_index=True, right_index=True)
```
