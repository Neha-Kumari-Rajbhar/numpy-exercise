```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "Category": ["Fruit", "Fruit", "Vegetable", "Vegetable", "Fruit"],
    "Item": ["Apple", "Banana", "Carrot", "Potato", "Mango"],
    "Price": [100, 40, 30, 20, 60],
    "Quantity": [5, 7, 3, 4, 6]
})

print(df)
```

```python
#Total price for each category
#df.groupby("Category")["Price"].sum()
df.groupby("Category")[["Price","Quantity"]].sum()
```

```python
#Average quantity for each category
df.groupby("Category")["Quantity"].mean()
```

```python
#Multiple aggregations at once
df.groupby("Category").agg({
    "Price":["sum","max","min"],
    "Quantity":["sum","mean","median"]
})
```

```python
df["Price"].mode().iloc[3]
```

```python
#for mode opreation
df.groupby("Category")["Price"].apply(lambda x: x.mode().iloc[0])
```

```python
#if you want the full list of modes per group
df.groupby("Category")["Price"].apply(lambda x: list(x.mode()))
```

```python
df["Quantity"].median()
```

```python
df2=pd.DataFrame({
    "City":["Delhi","Delhi","Delhi","Mumbai","Mumbai","Mumbai"],
    "Product":["Shoes","Shoes","Watch","Shoes","Watch","Watch"],
    "Sales":[200,150,300,400,250,300]
})
df2
```

```python
#Total sales of Shoes in each city
citydf=df2.groupby("City")["Sales"].sum()
citydf
```

```python
# Total sales of Watch in each city
prodf=df2.groupby("Product")["Sales"].sum()
prodf
```

```python
pd.concat([citydf,prodf]) #to solve this pivotTables used
```

```python
result = df2.groupby(["Product", "City"])["Sales"].sum()
result = result.unstack(fill_value=0)
print(result)
```

```python
g = df2.groupby([ "City","Product"])["Sales"].sum().unstack()
print(g)
```
